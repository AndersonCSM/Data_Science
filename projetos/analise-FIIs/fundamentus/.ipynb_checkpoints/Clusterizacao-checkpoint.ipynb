Neste arquivo iremos testar diferentes algoritmos de clusterização para obter diferentes resultados no agrupamentos dos fundos imobiliários a fim de encontrar um que se destaque dos demais

```python
# importando as bibliotecas para a análise
import pandas as pd
import numpy as np
import os
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.cluster import MiniBatchKMeans
```

```python
# Lendo um arquivo
dt = pd.read_csv("planilhas_fundamentus_fii/teste_31_07_2023.csv")
```

```python
# Modificando uma configuração do pandas para melhorar a visualização de valores muito grandes
pd.set_option('display.float_format', '{:.4f}'.format)
```

```python
# Visualizando os dados
dt
```

```python
dt_km = dt.copy()
```

```python
numeric = dt_km.drop(["papel","segmento"], axis=1)
```

```python
scaler = StandardScaler()
numeric= scaler.fit_transform(numeric)
```

# KMEANS - Não agrupou como esperado

```python
km = KMeans(n_clusters=28, init='k-means++', random_state=0)

km.fit(numeric)

dt_km['cluster_label'] = km.labels_

dt_km.cluster_label.hist()
```

```python
from time import sleep
for i in range(2, 20):
    km = KMeans(n_clusters=i, init='k-means++', random_state=0)

    km.fit(numeric)

    dt_km['cluster_label'] = km.labels_

    # dt_km.cluster_label.hist()
    
    print(dt_km.cluster_label[(dt_km.dividend_yield > 0.12)].unique())
    print(len(dt_km.cluster_label[(dt_km.dividend_yield > 0.12)].unique()))
```

```python
dt_km[(dt_km.dividend_yield > 0.12) & (dt_km.p_sobre_vp <= 1)]
```

```python
dt_km[dt_km.cluster_label == 18]
```

# MiniBatchMeans - Não agrupou como esperado

```python
from time import sleep
for i in range(2, 20):
    km = MiniBatchKMeans(n_clusters=i, init='k-means++', random_state=0, batch_size=3073)

    km.fit(numeric)

    dt_km['cluster_label'] = km.labels_

    # dt_km.cluster_label.hist()
    
    print(dt_km.cluster_label[(dt_km.dividend_yield > 0.12)].unique())
    print(len(dt_km.cluster_label[(dt_km.dividend_yield > 0.12)].unique()))
```

```python
km = MiniBatchKMeans(n_clusters=18, init='k-means++', random_state=0, batch_size=3073)

km.fit(numeric)

dt_km['cluster_label'] = km.labels_

# dt_km.cluster_label.hist()

print(dt_km.cluster_label[(dt_km.dividend_yield > 0.12)].unique())
print(len(dt_km.cluster_label[(dt_km.dividend_yield > 0.12)].unique()))
```

```python
dt_km.cluster_label.hist()
```

```python
dt_km[dt_km.cluster_label == 17]
```

# DBSCAN AGRUPOU COMO ESPERADO

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
db = DBSCAN(eps=0.30, min_samples=2)
db.fit(numeric)

dt_km['cluster_label'] = db.labels_

plt.hist(dt_km['cluster_label'], bins=len(set(db.labels_)))
plt.xlabel('Cluster Label')
plt.ylabel('Count')
plt.title('DBSCAN Cluster Labels Histogram')
plt.show()
```

```python
dt_km.cluster_label[dt_km.dividend_yield >= 0.12].unique()
```

```python
dt_km.cluster_label[(dt_km.dividend_yield >= 0.12) & (dt_km.ffo_yield >= dt_km.dividend_yield) & (dt_km.p_sobre_vp <= 1)].unique()
```

# AffinityPropagation Não agrupou como esperado

```python
from sklearn.cluster import AffinityPropagation
```

```python
aff = AffinityPropagation(damping=0.50, max_iter=900, random_state=0, convergence_iter=15) #definindo o modelo de clustering que iremos utilizar

aff.fit(numeric)

dt_km['cluster_label'] = aff.labels_

plt.hist(dt_km['cluster_label'], bins=len(set(aff.labels_)))
plt.xlabel('Cluster Label')
plt.ylabel('Count')
plt.title('DBSCAN Cluster Labels Histogram')
plt.show()
```
