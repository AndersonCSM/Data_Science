## Visualizações exploratórios usando o Pandas

```python
# Visualização de histograma
df_train.Fare.hist(bins=30)
# o ; serve para não mostrar o texto de saída, apenas o gráfico
```

```python
# Gráfico de barras para contagem de valores
df_train.Pclass.value_counts().plot.bar(grid=True)
```

```python
# Distribuição de densidade normal ou distribuição gaussiana para a colunas Fare
# kernel density estimation(KDE)
df_train["Fare"].plot.kde()
```
