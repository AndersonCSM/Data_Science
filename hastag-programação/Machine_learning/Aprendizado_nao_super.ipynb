# Aprendizado n√£o Supervisionado K-means

```python
# Importando o dataset
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True, as_frame=True)
```

```python
# Dataset
X = X.loc[:, ["petal length (cm)","petal width (cm)"]]
```

```python
# Visualizando os pontos
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(X["petal length (cm)"], X["petal width (cm)"])

plt.show()
```

```python
# Importando o Kmeans
from sklearn.cluster import KMeans
```

```python
# Criando o cluster e fazendo o treinamento
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

```python
# O 0 e 1 s√£o apenas r√≥tulos para os grupos, n√£o s√£o uma classifica√ß√£o
kmeans.labels_
```

```python
# Visualizando os pontos
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(X["petal length (cm)"], X["petal width (cm)"], c=kmeans.labels_)

plt.show()
```

```python
# Criando o cluster e fazendo o treinamento
kmeans2 = KMeans(n_clusters=3, random_state=0).fit(X)
kmeans3 = KMeans(n_clusters=4, random_state=0).fit(X)
kmeans4 = KMeans(n_clusters=5, random_state=0).fit(X)
kmeans5 = KMeans(n_clusters=6, random_state=0).fit(X)
```

```python
# Visualizando os pontos
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(10, 10))

ax[0, 0].scatter(X["petal length (cm)"], X["petal width (cm)"], c=kmeans.labels_)
ax[0, 1].scatter(X["petal length (cm)"], X["petal width (cm)"], c=kmeans2.labels_)
ax[1, 0].scatter(X["petal length (cm)"], X["petal width (cm)"], c=kmeans3.labels_)
ax[1, 1].scatter(X["petal length (cm)"], X["petal width (cm)"], c=kmeans4.labels_)
ax[2, 0].scatter(X["petal length (cm)"], X["petal width (cm)"], c=kmeans5.labels_)
ax[2, 1].remove()

plt.show()
```

```python
# Percorrendo diferentes valores de K

valores_k = []
inercias = []

for i in range(1,15):
    kmeans6 = KMeans(n_clusters=i, random_state=0).fit(X);
    inercias.append(kmeans6.inertia_)
    print(kmeans6.inertia_)
```

```python
# Visualizando os cotovelos para K grupos
fig, ax = plt.subplots()

ax.plot([x for x in range(1,15,1)], inercias)

ax.set(ylim=0, xlim=0)

plt.grid(alpha=0.3)
plt.show()
```

```python
# Verficando o adjusted_rand_score para k = 2
from sklearn import metrics

metrics.adjusted_rand_score(y, kmeans.labels_)
```

```python
# Para kmeans diferente
metrics.adjusted_rand_score(y, kmeans2.labels_)
```

```python
# Gerando uma lista de kmeans
ARI = [] # adjusted_rand_score
valores_k = []

for i in range(1,15):
    kmeans6 = KMeans(n_clusters=i, random_state=0).fit(X)
    valores_k.append(i)
    inercias.append(kmeans6.inertia_)
    ARI.append(metrics.adjusted_rand_score(y, kmeans6.labels_))
```

```python
# Visualizando graficamente os ARI
fig, ax = plt.subplots()

ax.plot([x for x in range(1,15,1)], ARI)

ax.set(ylim=0, xlim=0)

plt.grid(alpha=0.3)
plt.show()
```

```python
# Verificando a silhouette_score para k = 2
metrics.silhouette_score(X, kmeans2.labels_)
```

```python
# Gerando uma lista de kmeans
S = [] # silhouette_score
valores_k = []

for i in range(2,15):
    kmeans6 = KMeans(n_clusters=i, random_state=0).fit(X)
    valores_k.append(i)
    inercias.append(kmeans6.inertia_)
    S.append(metrics.silhouette_score(X, kmeans6.labels_))
```

```python
# Visualizando graficamente silhouette_score
fig, ax = plt.subplots()

ax.plot(valores_k, S)

ax.set(ylim=0, xlim=0)

plt.grid(alpha=0.3)
plt.show()
```
